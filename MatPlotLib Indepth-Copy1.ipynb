# Barcharts and Histograms

```python
x = [2,4,6,8,10]
y = [6,7,8,2,4]

#note here you use plt.bar
plt.bar(x,y, label = 'Bars1')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Graph\nCheck it out')
plt.legend()
  
```

```python
x = [2,4,6,8,10]
y = [6,7,8,2,4]

x2 =[1,3,5,7,9]
y2 =[4,3,7,4,2]
#note here you use plt.bar
plt.bar(x,y, label = 'Bars1')
plt.bar(x2,y2, label = 'Bars2')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Graph\nCheck it out')
plt.legend()
  
```

```python
#you can also customize the colors of your bargraphs


x = [2,4,6,8,10]
y = [6,7,8,2,4]

x2 =[1,3,5,7,9]
y2 =[4,3,7,4,2]

#note here you use plt.bar
#you can add the color element
#you can put the color name or hex-code #---
#note k=black since b=blue
plt.bar(x,y, label = 'Bars1',color='b')
plt.bar(x2,y2, label = 'Bars2', color='k')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Graph\nCheck it out')
plt.legend()
```

# Scatter Plots

```python
x = np.arange(1,11)
y = np.random.randint(1,20,10)

plt.scatter(x,y, color='c',label='Skitskat') 

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')
plt.legend()
 
```

```python
#you can also alter the type of markers, the default is o
plt.scatter(x,y, color='c',label='Skitskat', marker='*') 
plt.legend()
```

```python
#you can also change the size of the markers
plt.scatter(x,y, color='c',label='Skitskat', marker='*',s=110) 
plt.legend()
```

# Stack Plots

```python
import matplotlib.pyplot as plt

days = np.arange(1,6)
sleeping =[7,8,6,11,7]
eating =[2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days, sleeping, eating, working, playing, colors =['m','c','r','k'])

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')

 
```

```python

days = np.arange(1,6)
sleeping =[7,8,6,11,7]
eating =[2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

#to input labels in stackplot, you create empty plots
plt.plot([],[], color = 'm', label= 'sleeping', linewidth=5)
plt.plot([],[], color = 'c', label= 'Eating',linewidth=5)
plt.plot([],[], color = 'r', label= 'working',linewidth=5)
plt.plot([],[], color = 'k', label= 'playing', linewidth=5)

plt.stackplot(days, sleeping, eating, working, playing, colors =['m','c','r','k'])

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')
plt.legend()
 
```

# Pie Charts

```python


slices= [7,2,2,13]
activities = ('sleeping eating working playing').split()

plt.pie(slices, labels = activities, colors = ('c r  m k ').split())

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')
plt.legend()
 
```

```python
slices= [7,2,2,13]
activities = ('sleeping eating working playing').split()

#you can also set the start angle
plt.pie(slices, labels = activities, colors = ('c r  m k ').split(), startangle=90)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')
plt.legend()
```

```python
slices= [7,2,2,13]
activities = ('sleeping eating working playing').split()
#we can add shadow to it
plt.pie(slices, labels = activities, colors = ('c r  m k ').split(), startangle=90, shadow=True)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')
plt.legend()
```

```python
slices= [7,2,2,13]
activities = ('sleeping eating working playing').split()
#we can pull a piece of the pie outwards using explode
plt.pie(slices, labels = activities, colors = ('c r  m k ').split(), 
        startangle=90, 
        shadow=True,
        explode=(0,0,0.1,0) )

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')
plt.legend()
```

```python
slices= [7,2,2,13]
activities = ('sleeping eating working playing').split()
#we can add the percentages on your pie chart using %1.1f%%
plt.pie(slices, 
        labels = activities, 
        colors = ('c r  m y ').split(), 
        startangle=90, 
        shadow=True,
        explode=(0,0,0.1,0),
        autopct = '%1.1f%%')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Title')
plt.legend()
```
